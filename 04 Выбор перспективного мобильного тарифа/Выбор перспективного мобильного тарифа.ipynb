```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
import seaborn as sns
import matplotlib.ticker as ticker
from scipy import stats as st
import math
```

 # ***Используемые функции***

1. **будет выводить следующие данные:**
  - info()
  - количество строк и столбцов
  - наличие пропусков
  - наличие дубликатов

```python
def main_charact(row):
        return row.info(),\
               row.shape,\
               row.isna().sum(),\
               row.duplicated().sum()
```

2. **Для переноса блока тарифов ultra, чтобы он стоял не снизу, а справа от блока тарифов smart. Чтобы составить процентные изменения**

```python
def change_table(row):
    a = row.query('tariff == "ultra"')   # отбираем строки с тарифом ultra
    b = row.query('tariff == "smart"')   # отбираем строки с тарифом smart
    c = b.merge(a,
                on='month',
                how='left',
                suffixes=('_smart', '_ultra'))  # прилепим ultra справа, добавляем подписи для различия
    return c
```

3. **будет выводить следующие данные:**
  - метод describe()
  - boxplot
  - левый и правый границы диаграммы размаха

```python
def analysis(row):
    q1 = row.quantile(0.25)
    q3 = row.quantile(0.75)
    iqr = q3 - q1
    fence_low = q1-1.5*iqr
    fence_high = q3+1.5*iqr     
    return row.describe(), sns.boxplot(row), fence_low, fence_high
```

4. **Для просмотра двух графиков на одной плоскости**

```python
def plot(row1, row2, y, title):
    ax = (row1
          .plot(kind='line',
                x='month',
                y=y,
                linewidth=4,
                alpha=0.7,
                color='green',
                figsize=(15,4))
         )

    (row2
          .plot(kind='line',
                linewidth=4,
                x='month',
                y=y,
                alpha=0.7,
                ax=ax,
                grid=True,
                legend=True,
                figsize=(15,4))
    )
    ax.xaxis.set_major_locator(ticker.MultipleLocator(1))
    plt.legend(['smart', 'ultra'])
    plt.title(title);
```

5. **Для просмотра процентных отношений на графике**

```python
def plot_perc(row, title):
    (row
          .plot(kind='line',
                linewidth=4,
                x='month',
                grid=True,
                legend=True,
                figsize=(15,4))
    )
    plt.title(title);
```

6. **Для вывода сравнительного графика средних показателей**

```python
def graph(row1,row2,y,title):
    ax = (row1
          .plot(kind='hist',
                y=y, 
                histtype='step',
                linewidth=5,
                alpha=0.7,
                color='green',
                figsize=(15,4),
                bins=12)
         )

    (row2
          .plot(kind='hist', 
                linewidth=4,
                y=y, 
                histtype='step', 
                alpha=0.7, 
                ax=ax, 
                grid=True,
                legend=True,
                figsize=(15,4),
                bins=12)
    )
    plt.legend(['smart', 'ultra'])
    plt.title(title);
```

**Примечание:**
«Мегалайн» всегда округляет секунды до минут, а мегабайты — до гигабайт. Каждый звонок округляется отдельно: даже если он длился всего 1 секунду, будет засчитан как 1 минута.
Для веб-трафика отдельные сессии не считаются. Вместо этого общая сумма за месяц округляется в бо́льшую сторону. Если абонент использует 1025 мегабайт в этом месяце, с него возьмут плату за 2 гигабайта.


- **Необходимо округлить продолжительность вызовов в большую сторону, при этом, если продолжительность меньше минуты, округляем до 1й минуты**

```python
df_cal.head()
```

```python
df_cal['duration'] = np.ceil(df_cal['duration'])
```

```python
df_cal['month'] = df_cal['call_date'].dt.month
```

```python
df_cal_pivot.head()
```

- ***Уберем верхний уровень названия колонок***

```python
df_cal_pivot.columns = [tup[1] if tup[1] else tup[0] for tup in df_cal_pivot.columns]
```

```python
df_cal_pivot = df_cal_pivot.reset_index()
```

```python
df_cal_pivot.head()
```

## **Проверка гипотез**

###  Средняя выручка пользователей тарифов «Ультра» и «Смарт» различаются

- ***Нулевая гипотеза - средняя выручка пользователей тарифов Ультра и Смарт одинаковая***
- ***Альтернативная гипотеза - средняя выручка пользователей тарифов Ультра и Смарт различны***

- ***Сделаем две выборки по тарифам. Данные выборки будут независимы друг от друга, тк тарифы не влияют друг на друга***

- ***Используем метод о равенстве среднего двух генеральных совокупностей по взятым из них выборкам***

```python
df_all_revenue_hypot = (df_all
                        .groupby(['tariff', 'user_id'])['revenue_from_user']
                        .mean()
                        .reset_index()
                       )
```

```python
df_all_revenue_hypot.sample(5)
```

```python
df_all_revenue_smart = df_all_revenue_hypot.query('tariff == "smart"')['revenue_from_user']
```

```python
df_all_revenue_ultra = df_all_revenue_hypot.query('tariff == "ultra"')['revenue_from_user']
```

- ***Критический уровень статистической значимости зададим 5%***

```python
alpha = .05 
```

```python
results = st.ttest_ind(
    df_all_revenue_smart, 
    df_all_revenue_ultra)
print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

- *Средняя выручка пользователей тарифов «Ультра» и «Смарт» различаются. Так и есть. Средняя прибыль тарифа СМАРТ выше, те тариф УЛЬТРА имеет излишек бесплатных услуг*

### Средняя выручка пользователей из Москвы отличается от выручки пользователей из других регионов.

- ***Альтернативная гипотеза - средняя выручка пользователей из Москвы и из других регионов различны***
- ***Сделаем две выборки по городам. Данные выборки будут независимы друг от друга, тк пользователи не влияют друг на друга***
- ***Используем метод о равенстве среднего двух генеральных совокупностей по взятым из них выборкам***

```python
df_all_revenue_hypot = (df_all
                        .groupby(['city', 'user_id'])['revenue_from_user']
                        .mean()
                        .reset_index()
                       )
```
