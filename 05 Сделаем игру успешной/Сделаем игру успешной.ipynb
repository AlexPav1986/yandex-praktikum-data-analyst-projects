                - Обнаружено 269 пропусков

            	- Как исправлять: Подставим год из названия игр, далее заполним соседними значениями.
            	- Необходимый тип: переведем в целочисленные значения

- **У некоторых игр год указан в конце названия.**
- **Возьмем последние 4 символа названия, поместим в новый столбец**
- **Если 4 символа будут числом, то это будет годом релиза, заменим соответствующий пропуск**
- **Далее заменим тип столбца на integer**

```python
df['year_from_name_games'] = df['name'].map(lambda x : str(x)[-4:])
```

```python
df[df['year_of_release'].isna()].head()
```

- ***Сохраним в новую переменную данную выборку с пропусками***

```python
year_of_release_nan = df[df['year_of_release'].isna()]
```

- ***Пройдемся циклом по выборке***

```python
import warnings

warnings.filterwarnings("ignore")


for i in range(len(year_of_release_nan['year_of_release'])):
    
    # если в ячейке year_from_name_games число - меняем пропуск в year_of_release на это число
    
    if year_of_release_nan['year_from_name_games'].iloc[i].isdigit() == True:
        year_of_release_nan['year_of_release'].iloc[i] = year_of_release_nan['year_from_name_games'].iloc[i]
```

```python
year_of_release_nan.head()
```

```python
year_of_release_nan['year_of_release'].unique()
```

- ***Теперь заменим строки в df на строки из year_of_release_nan с замененными пропусками***

```python
df.loc[year_of_release_nan.index] = year_of_release_nan.loc[year_of_release_nan.index]
```

```python
df.loc[year_of_release_nan.index].head()
```

- ***Проверим количество пропусков в данном столбце***

```python
df = df.drop(columns = ['year_from_name_games'],axis = 1)
```

### топ-5 платформ по регионам

```python
def top_5_plat_on_reg(region_sales):
    
    # будет выводить общую прибыль по региону,
    # ТОП-5 по региону
    # и небольшую визуализацию
    
    share_plat = (df_actual
                        .groupby('platform')[region_sales]
                        .sum()
                        .sort_values(ascending=False)
                        .head()
                        .reset_index()
                 )

    plt.pie(share_plat[region_sales], labels=share_plat['platform'])

    # add a circle at the center to transform it in a donut chart
    my_circle=plt.Circle( (0,0), 0.7, color='white')
    p=plt.gcf()
    p.gca().add_artist(my_circle)
   
    print(f'Общая прибыль в регионе {region_sales} млн - {int(df_actual[region_sales].sum())}  млн')
    display(share_plat)
```

```python
#fig = plt.figure()
fig = plt.figure(figsize=(15,5))

ax_1 = fig.add_subplot(1, 3, 1)
top_5_plat_on_reg('na_sales')
    
ax_2 = fig.add_subplot(1, 3, 2)
top_5_plat_on_reg('eu_sales')

ax_3 = fig.add_subplot(1, 3, 3)
top_5_plat_on_reg('jp_sales')

ax_1.set(title = 'na_sales')
ax_2.set(title = 'eu_sales')
ax_3.set(title = 'jp_sales');
```

### топ-5 жанров по регионам

- **Используем средние продажи, тк некоторые игры могут принести перекос**

```python
def top_5_gen_on_reg(region_sales):
    
    # выводит ТОП-5 по региону
    
    share_gen = (df_actual
                         .groupby('genre')[region_sales]
                         .mean()
                         .sort_values(ascending=False)
                         .head()
                         .reset_index()
                 )
    
    plt.pie(share_gen[region_sales], labels=share_gen['genre'])
    # add a circle at the center to transform it in a donut chart
    my_circle=plt.Circle( (0,0), 0.7, color='white')
    p=plt.gcf()
    p.gca().add_artist(my_circle)
        
    display(share_gen)
```

```python
fig = plt.figure(figsize=(15,5))

ax_1 = fig.add_subplot(1, 3, 1)
top_5_gen_on_reg('na_sales')
    
ax_2 = fig.add_subplot(1, 3, 2)
top_5_gen_on_reg('eu_sales')

ax_3 = fig.add_subplot(1, 3, 3)
top_5_gen_on_reg('jp_sales')

ax_1.set(title = 'na_sales')
ax_2.set(title = 'eu_sales')
ax_3.set(title = 'jp_sales');
```
